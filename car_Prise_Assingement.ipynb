```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Print and read csv file
dff = pd.read_csv("CarPrice_Assignment.csv")
df  = dff.copy()
```

```python
#briefing
df.info()
```

```python
#SUM OF NULL Values
df.isna().sum()
```

```python
sns.heatmap(df.isna())
plt.show()
```

```python
#CHECKING NULL PERSENTAGE
df.isna().sum()/len(df)*100
```

```python
#check dublicate
df.duplicated()
```

```python
#Drop Null values
df.dropna()
```

```python
#Drop unwanter columns
df =df.drop(columns = "aspiration")
df =df.drop(columns = "doornumber")
df =df.drop(columns = "carbody")
df =df.drop(columns = "drivewheel")
df =df.drop(columns = "car_ID")
df
```

```python
#Scatter polot
df.plot.scatter(x='citympg', y='highwaympg' , title = "Car ratio" )
plt.show
```

```python
#Correction in missing values
df["CarName"] = df["CarName"].str.replace("alfa-romero giulia","Ranoult Duster ")
df["CarName"] = df["CarName"].str.replace("alfa-romero stelvio"," kia Seltos")
df["CarName"] = df["CarName"].str.replace("volvo","Volvo")
df["CarName"] = df["CarName"].str.replace("audi","Audi")
df
```

```python
#Converting values into boolean or integar value
df['fueltype'] = df['fueltype'].map({'gas':1,'diesel':0} )
df['enginelocation'] = df['enginelocation'].map({'front':1,'rear':0} )
df['fuelsystem'] = df['fuelsystem'].map({'mpfi':1,'idi':0} )
df
```

```python
#Bar plot
sns.barplot(x = "peakrpm",y = "price" , data = df)
plt.show()
```

```python
#Histogram using prise component
sns.displot(df["price"])
plt.show()
```
